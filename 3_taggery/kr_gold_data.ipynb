```python
import pandas as pd
from xml.dom import minidom
import xml.etree.ElementTree as ET
```

```python
xmldoc = minidom.parse('../Dane/gold-task-a-b.xml')
itemlist = xmldoc.getElementsByTagName('orth')
itemlist = list(map(lambda x: x.firstChild.nodeValue, itemlist))
itemlist = ' '.join(itemlist)
with open('../Dane/raw.txt', 'x') as f:
    f.write(itemlist)
```

```python
def ccl_bases_tags(tree):
    return {tok.find('./lex/base').text: tok.find('./lex/ctag').text for tok in tree.iter('tok')}
```

```python
tree = ET.parse('../Dane/gold-task-a-b.xml')
words = ccl_bases_tags(tree)
```
